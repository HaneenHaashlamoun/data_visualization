```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
sns.barplot(x='petal_length', y='petal_width', data=df)
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=df)
```

```python
sns.regplot(x='sepal_length', y='petal_length', data=df, color='purple')
```
