```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

- Optionally use dataset built into Seaborn
- Load the data you receive into a Pandas DataFrame

```python
df = sns.load_dataset('car_crashes')
# help(sns.load_dataset)
df.head()
```

```python
sns.barplot(x='speeding', y='total', data=df)
```

```python
sns.lineplot(x='speeding', y='total', data=df)
```

```python
sns.regplot(x='speeding', y='total', data=df, color='blue')
```
